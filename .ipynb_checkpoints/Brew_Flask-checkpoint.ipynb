```python
 %matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import datetime as dt
```

```python
 # Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect, MetaData, Table, Column, Integer, String, Float, cast
```

```python
 # create engine to brewBuds.sqlite
engine = create_engine("sqlite:///brewBuds.sqlite")
```

```python
metadata = MetaData()
```

```python
Table('brewery', metadata,
     Column('Index', Integer, primary_key = True),
     Column('City', String),
     Column('Latitude',Float),
     Column('Longitude',Float),
     Column('Name',String),
     Column('PostalCode',Integer),
     Column('State', String))
```

```python
Table('starbucks', metadata,
     Column('Index', Integer, primary_key = True),
     Column('State', String),
     Column('PostalCode',Integer),
     Column('Longitude',Float),
     Column('Latitude',Float)
     )
```

```python
Table('income', metadata,
     Column('Index', Integer, primary_key = True),
     Column('PostalCode',Integer),
     Column('Median_Income',Float),
     Column('Population',Float),
     Column('Median_age',Float) 
     Column('Keydemo_per',Float)
     )
```

```python
 # reflect an existing database into a new model
Base = automap_base(metadata=metadata)
# reflect the tables
Base.prepare()
```

```python
 # View all of the classes that automap found
Base.classes.keys()
```

```python
 # Save references to each table
brewery = Base.classes.brewery
starbucks = Base.classes.starbucks
income = Base.classes.income
```

```python
 # Create our session (link) from Python to the DB
session = Session(engine)
```

```python
# number of breweries in each zipcode
brew_count_zip = session.query(brewery.PostalCode, func.count(brewery.Index)).\
    group_by(brewery.PostalCode).\
    order_by(func.count(brewery.Name).desc()).all()
for i in brew_count_zip:
    print(i)
```

```python
# number of breweries in each zipcode combined with median income
sel = [brewery.PostalCode, func.count(brewery.Name), income.Population, income.Median_Income, income.PerUnder60]
bi_join = session.query(*sel).\
group_by(brewery.PostalCode).\
filter(brewery.PostalCode ==income.PostalCode).limit(10).all()

for record in bi_join:
    print(record)
#         f"The European animal '{ea_fam} {ea_gen} {ea_spec}'"
#         f"belongs to the same sporder as the North American animal '{na_fam} {na_gen} {na_spec}'.")
```
